# Using the Otto theme

```python
t = np.arange(0,10, step=0.01)
y = np.cos(t)
```

```python
df = pd.DataFrame({'t':t, 'y':y})
```

```python
ggplot(df, aes(x='t', y='y'))+geom_line(color=otto_red)+theme_otto()
```

```python
sin_t = np.sin(t)
cos_t = np.cos(t)
df = pd.DataFrame({'t':t, 'sin':sin_t, 'cos':cos_t})
df = df.melt(id_vars='t', var_name='function', value_name='y')
df.head()
```

```python
ggplot(df, aes(x='t', y='y', color='function'))+geom_line()+scale_color_manual(otto_colors)+theme_otto()
```

```python
ggplot(df, aes(x='t', y='y'))+geom_line(color=otto_red)+theme_otto()+facet_wrap('function', nrow=2)
```

# Using theme_xkcd
